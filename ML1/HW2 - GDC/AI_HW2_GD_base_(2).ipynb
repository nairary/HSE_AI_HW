## Задание 2 (1 балл)

Найдите экстремумы этих функций в Python.

```python
import sympy as sp
import numpy as np
from matplotlib import pyplot as plt

x = sp.Symbol('x')

def f(x):
  return x**2 - 30*x + 1

f_prime = sp.diff(f(x))
crit_points_f = sp.solve(f_prime, x)
print(f_prime, crit_points_f)

def g(x):
  return -x**3 + 5*x - 7*x**2 + 17

g_prime = sp.diff(g(x))
crit_points_g = sp.solve(g_prime, x)
print(g_prime, crit_points_g)

def h(x):
  return x**2 - 1/x

h_prime = sp.diff(h(x))
crit_points_h = sp.solve(h_prime, x)
print(h_prime, crit_points_h)

def j(x):
  return 1/x**5 + x**17 + 8
j_prime = sp.diff(j(x))
crit_points_j = sp.solve(j_prime, x)
print(j_prime, crit_points_j)
```

```python
import matplotlib.pyplot as plt

# Не могу использовать код для отображения графика из семинара, потому что получаю ошибки в точках где делю на ноль
# буду использовать sp
f_sp_foo = x**2 - 30*x + 1
g_sp_foo = -x**3 + 5*x - 7*x**2 + 17
h_sp_foo = x**2 - 1/x
j_sp_foo = 1/x**5 + x**17 + 8

def plotter(x_start, x_end, y_start, y_end, foo, title, crit_points, bend_points):
  x_values = np.linspace(x_start, x_end, 1000)
  foo_values = sp.lambdify(x, foo, 'numpy')(x_values)

  plt.axvline(x=crit_points[0], c = 'black')
  plt.axhline(y=foo.subs(x, float(crit_points[0])), c = 'black')

  plt.xlim([x_start, x_end])
  if y_start != y_end:
    plt.ylim([y_start, y_end])

  plt.plot(x_values, foo_values)
  plt.title(f'График функции {title}(x) c экстремумами и точками перегиба')

  for point in crit_points:
    if sp.im(point) == 0:
      plt.scatter(float(point), foo.subs(x, float(point)), color='red', label='Экстремум')
      plt.annotate(f'{float(point):.2f}',
                         (float(point), foo.subs(x,crit_points[0])),
                         textcoords='offset points', xytext=(0, -12),
                         ha='left', color='black')

  for point in bend_points:
    if sp.im(point) == 0:
      plt.scatter(float(point), foo.subs(x, float(point)), color='blue', label='Точка перегиба')
      plt.annotate(f'{float(point):.2f}',
                      (float(point), foo.subs(x,crit_points[0])),
                      textcoords='offset points', xytext=(0,-12),
                      ha='left', color='blue')

  plt.xlabel('x')
  plt.ylabel(f'{title}(x)')
  plt.show()

plotter(10, 20, -225, -200, f_sp_foo, 'f', crit_points_f, bend_points_f)
plotter(-7, 2, -65, 25, g_sp_foo, 'g', crit_points_g, bend_points_g)
plotter(-10, 10, -30, 30, h_sp_foo, 'h', crit_points_h, bend_points_h)
plotter(-5, 5, -20, 20, j_sp_foo, 'j', crit_points_j, bend_points_j)
```

## Задание 6 (1 балл)

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
def print_gdc(x_curr, trace, x_start, x_end, foo, title):
  x_values = np.linspace(x_start, x_end, 1000)
  f_values = sp.lambdify(x, foo, 'numpy')(x_values)

  plt.plot(x_values, f_values)

  plt.title(f'График функции {title}(x) c экстремумами и точками перегиба')

  plt.xlabel('x')
  plt.ylabel(f'{title}(x)')

  trace_values = [foo.subs(x, point) for point in trace]
  plt.scatter(trace, trace_values, c='red')

  print(f'Начальное приближение: {trace[0]}')
  print(f'Локальный минимум: {round(x_curr, 5)}')
  plt.show()

def gradient_descent(x_start_gdc, epsilon, num_iterations, x_start, x_end, foo, title, learning_rate=0.1):
  print(f'learning_rate: {learning_rate}')
  x_curr = x_start_gdc
  df_x = sp.diff(foo, x) # производная

  trace = []
  trace.append(x_curr)

  for i in range(num_iterations):
    x_new = x_curr - learning_rate * df_x.subs(x, x_curr)
    trace.append(x_new)

    if abs(x_new - x_curr) < epsilon:
      return print_gdc(x_curr, trace, x_start, x_end, foo, title)

    x_curr = x_new

  return print_gdc(x_curr, trace, x_start, x_end, foo, title)


gradient_descent(10, 1e-5, 100, 10, 20, f_sp_foo, 'f')
gradient_descent(-7, 1e-5, 100, -7, 2, g_sp_foo, 'g', learning_rate=0.01)
gradient_descent(1, 1e-5, 100, -10, 10, h_sp_foo, 'h')
gradient_descent(0.7, 1e-5, 100, 0.5, 1.2, j_sp_foo, 'j', learning_rate=0.001)
```

## Задание 8 (1.5 балла)

Реализуйте [метод Ньютона](https://en.wikipedia.org/wiki/Newton%27s_method_in_optimization) (метод второго порядка) для поиска минимума функции.

Про метод ещё можно почитать [здесь](https://math.semestr.ru/optim/minimum.php).

```python
def print_newthon(x_curr, trace, x_start, x_end, foo, title):
  x_values = np.linspace(x_start, x_end, 1000)
  f_values = sp.lambdify(x, foo, 'numpy')(x_values)

  plt.plot(x_values, f_values)

  plt.title(f'График функции {title}(x) c экстремумами и точками перегиба')

  plt.xlabel('x')
  plt.ylabel(f'{title}(x)')

  trace_values = [foo.subs(x, point) for point in trace]
  plt.scatter(trace, trace_values, c='red')

  print(f'Начальное приближение: {trace[0]}')
  print(f'Локальный минимум: {round(x_curr, 5)}')
  plt.show()

def newton_method(x_curr, epsilon, max_iter, x_start, x_end, foo, title):

    df_x = sp.diff(foo, x)
    d2f_x = sp.diff(df_x, x)

    df_func = sp.lambdify(x, df_x, 'numpy')
    d2f_func = sp.lambdify(x, d2f_x, 'numpy')

    trace = []
    trace.append(x_curr)

    try:
        for i in range(max_iter):
            clipped_x_curr = np.clip(x_curr, 1e-5, None) if title == 'h' else x_curr
            df_val = df_func(clipped_x_curr)
            d2f_val = d2f_func(clipped_x_curr)

            if d2f_val == 0:
                raise ValueError("df2 is zero")

            delta = -df_val / d2f_val
            print(delta)
            x_next = x_curr + delta

            trace.append(x_next)

            print(f"i = {i+1}: x = {x_next}, f(x) = {foo.subs(x, x_next)}")

            if abs(delta) < epsilon:
                print_newthon(x_curr, trace, x_start, x_end, foo, title)
                return

            x_curr = x_next

        print_newthon(x_curr, trace, x_start, x_end, foo, title)

    except ValueError as e:
        print(e)
```
